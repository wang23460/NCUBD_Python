#### 用圖片來看為什麼需要**正規化**
房子數量再沒有正規化的情況下，看起來數字根本一樣

```python
house = np.array([3, 0, 1, 10, 0,])
salary = np.array([0, 1000000, 700000, 10000000, 200000])
plt.xlim(0, np.max(salary)) # 設定一下座標區間，因為 plot 預設會顯示該欄位最大到最小這個區間（等同正規化了），這樣的話 x.y 軸基準不一樣，無法比較，也無法顯示真時期況
plt.scatter(house, salary)
```

把兩個屬性都正規化到 0-1 之間
* (x - min) / (max - min)

```python
house = (house - np.min(house)) / (np.max(house) - np.min(house))
salary = (salary - np.min(salary)) / (np.max(salary) - np.min(salary))
plt.scatter(house, salary)
```

#### plot有兩種呈現方式
以 [2, 1, 5] 為例（下面的名詞是我自己掰的）:

| 類型 | 描述 | 函數 |
| ------:| -----------:| -----------:|
| 統計式   | 第一筆資料的 x 軸位置是 2 | plot、scatter... |
| 圖片式 | 第一個位置的顏色是遣灰色（以255為例） | imshow |




```python
# 這裡不重要，只是要生出樹的陣列而已
def create_tree_array(tree_width, trunk_width=1, trunk_height=1):
  tree_height = int(tree_width / 2 + 1)
  tree = []
  center = int(tree_width / 2)
  for i in range(tree_height):    
    layers = [ 0 if j < center-i or j > center+i else 1 for j in range(tree_width) ]
    tree.append(layers)
  
  half_trunk_width = int(trunk_width / 2)
  for _ in range(trunk_height):
    tree.append([ 0 if j < center-half_trunk_width or j > center+half_trunk_width else 1 for j in range(tree_width) ])
  return tree
```

```python
tree = create_tree_array(13)
tree
```

```python
import random
def create_tree_array_with_decoration(tree_width, trunk_width, trunk_height):
  tree = create_tree_array_with_decoration(tree_width, trunk_width, trunk_height)
  tree_height = int(tree_width/2) + 1
  leaves_amount = tree_width * tree_height
  print(len(tree),len(tree[0]))
  for _ in range(random.randint(0,leaves_amount)):
    select_width = random.randint(0, tree_width-1)
    select_height = random.randint(0, tree_height-1)
    if tree[select_height][select_width] == 1:
      tree[select_height][select_width] = random.random()
  
  return tree
```
