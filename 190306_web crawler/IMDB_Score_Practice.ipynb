```python
from urllib.request import urlopen
from bs4 import BeautifulSoup

html = urlopen("https://ppt.cc/fx3k8x")
bsObj = BeautifulSoup(html,'html.parser')
namelist = bsObj.findAll("td",{"class":"titleColumn"})
scorelist = bsObj.findAll("td",{"class":"ratingColumn imdbRating"})
```

```python
# 爬出熱門電影名稱及其分數
print('IMDB Most Popular 100 Movie with Score')
for a in range(0,1):
    for x,y in zip(namelist,scorelist):
        a+=1
        b=str(a)
        testa = []    
        testa = x.get_text().splitlines()
        testb = []
        testb = y.get_text().splitlines()
        name_score = testa[1] + ':' + testb[-1]    
        print(b+". "+ name_score)
```
